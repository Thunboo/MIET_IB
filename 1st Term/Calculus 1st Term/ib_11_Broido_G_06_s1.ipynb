```python
#  1
y1 = (2*x - 1) / (x - 1)
y2 = (2*x - 1) / abs(x - 1)

sp.plot((y1, (x, -5, 1)), (y1, (x, 1, 2)), (y1, (x, 2, 5)))
print("lim(x->+inf) (y1) = 2")
print("lim(x->-inf) (y1) = 2")
print("lim(x->1+0) (y1) = +inf")
print("lim(x->1-0) (y1) = -inf")
print("lim(x->1) (y1) = inf")
print("lim(x->2+-0) (y1) = 3")
sp.plot((y2, (x, -5, 1)), (y2, (x, 1, 2)), (y2, (x, 2, 5)))
print("lim(x->+inf) (y2) = 2")
print("lim(x->-inf) (y2) = -2")
print("lim(x->1+0) (y2) = +inf")
print("lim(x->1-0) (y2) = +inf")
print("lim(x->1) (y2) = +inf")
print("lim(x->2+-0) (y2) = 3")
```

```python
#  1
y1 = sp.Abs( (2*x - 1) / (x - 1) )
y2 = sp.Abs(3*x - 6) / (x - 2)
sp.plot(y1, (x, 1.5, 2.5))
sp.plot((y1, (x, -5, 1)), (y1, (x, 1, 7)))
print("lim(x->+inf) (y1) = 2")
print("lim(x->-inf) (y1) = 2")
print("lim(x->1+-0) (y1) = +inf")
print("lim(x->2+-0) (y1) = 3")
sp.plot((y2, (x, -8,2)), (y2, (x, 2, 12)))
print("lim(x->+inf) (y2) = 3")
print("lim(x->-inf) (y2) = -3")
print("lim(x->1+-0) (y2) = -3")
print("lim(x->2-0) (y2) = -3")
print("lim(x->2+0) (y2) = 3")
```
