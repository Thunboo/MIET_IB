```python
import numpy as np 
import matplotlib.pyplot as plt
import math

OA = np.array([-2,0])
OB = np.array([1, 2])
OC = np.array([1,-1])

print('Coordinates:')
AB = OB - OA
BC = OC - OB
AC = OC - OA

print('AB:', AB)
print('BC:', BC)
print('AC:', AC)

plt.grid()
plt.ylim(-2,3)
plt.xlim(-3,2)
plt.arrow(-2,0, AB[0], AB[1], width = 0.07, 
          length_includes_head = True, #Cтрелка ровно по вектору
          facecolor = 'Yellow')
plt.arrow(1,2, BC[0], BC[1], width = 0.07, 
          length_includes_head = True, 
          facecolor = 'Red')
plt.arrow(-2,0, AC[0], AC[1], width = 0.07, 
          length_includes_head = True, 
          facecolor = 'Blue')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import math

OA = np.array([-2,0])
OB = np.array([1, 2])
OC = np.array([1,-1])

AB = OB - OA
BC = OC - OB
AC = OC - OA
AD = AC - AB
CD = BC - AD - AB

print(AB)
plt.grid()
plt.ylim(-4,3)
plt.xlim(-4,3)
plt.arrow(-2,0, AB[0], AB[1], width = 0.07, 
          length_includes_head = True, #Cтрелка ровно по вектору
          facecolor = 'Yellow')

plt.arrow(1,2, BC[0], BC[1], width = 0.07, 
          length_includes_head = True, 
          facecolor = 'Red')

plt.arrow(1,-1, CD[0], CD[1], width = 0.07, 
          length_includes_head = True, 
          facecolor = 'Blue')

plt.arrow(-2,-3, DA[0], DA[1], width = 0.07, 
          length_includes_head = True, 
          facecolor = 'Orange')
plt.show()
```

```python
# 3Dimensional vector graphing
import matplotlib.pyplot as plt
import numpy as np
import math

a = np.array([0, 5, -1])
b = np.array([-4, 9, 3])

cos_angle = np.dot(a, b) / ( np.linalg.norm(a) * np.linalg.norm(b) )
angle = np.arccos(cos_angle)
print(angle, 'rads')
angle = math.degrees(angle)
print(angle, 'degrees')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#%matplotlib ipympl
#%matplotlib widget

ax = plt.axes(projection='3d')

z = np.array([0,0,0])
a = np.array([1,2,0])
b = np.array([2,1,0])
c = np.cross(a,b)   # Multiplication

ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.set_zlim([-2,2])

ax.quiver(*z, *a, color='Blue') # * = распаковать.  quiver - вектор
ax.quiver(*z, *b, color='Green')
ax.quiver(*z, *c, color='Red')

ax.axis()
plt.grid()
plt.show()
```

```python
# ДОМАШНЕЕ ЗАДАНИЕ
# № 1
import numpy as np
import matplotlib.pyplot as plt
import math

ax = plt.axes(projection='3d')

z = np.array([0,  0,  0])
A = np.array([1,  3, -1])
B = np.array([2, -1,  4])
C = np.array([5,  0,  3])
AB = B - A
BC = C - B
AC = C - A

#c = np.cross(a,b)   # Multiplication
#print(AB, BC, AC)

ax.set_xlim([-4,6])
ax.set_ylim([-4,6])
ax.set_zlim([-4,6])

ax.quiver(*A, *AB, color='Blue')
ax.quiver(*B, *BC, color='Green')
ax.quiver(*A, *AC, color='Red')

#Calculating an area of a triangle
Matrix_x = np.matrix([[1, A[1], A[2]],
                      [1, B[1], B[2]],
                      [1, C[1], C[2]]])
S_x = 0.5 * np.linalg.det(Matrix_x)

Matrix_y = np.matrix([[A[0], 1, A[2]],
                      [A[0], 1, B[2]],
                      [A[0], 1, C[2]]])
S_y = 0.5 * np.linalg.det(Matrix_y)

Matrix_z = np.matrix([[A[0], A[1], 1],
                      [B[0], B[1], 1],
                      [C[0], C[1], 1]])
S_z = 0.5 * np.linalg.det(Matrix_z)

S = math.sqrt( S_x**2 + S_y**2 + S_z**2 )
print("Area of a triangle equals to ~", S)

ax.axis()
plt.grid()
plt.show()
```
