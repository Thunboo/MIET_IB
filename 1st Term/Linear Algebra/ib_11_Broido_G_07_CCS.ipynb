```python
import numpy as np
import matplotlib.pyplot as plt
# A(2, -4) в повернутой сис-ме, угол = 45 градусов
x1, y1 = 2*np.sqrt(3), -np.sqrt(3)
x0, y0 = -1, -2
alpha1 = np.pi / 6
def get_regular_coord(x_1,y_1, angle):
    x = x_1 * np.cos(angle) - y_1 * np.sin(angle) - 1
    y = y_1 * np.cos(angle) + x_1 * np.sin(angle) - 2
    return (x,y)
x, y = get_regular_coord(x1,y1, alpha1)

plt.arrow(-1,0,6,0, width=0.05, color='black')
plt.arrow(0,-2,0,7, width=0.05, color='black')

x3,y3 = np.cos(-alpha1), -np.sin(-alpha1)
x4,y4 = np.sin(-alpha1),np.cos(-alpha1)

plt.arrow(x0,y0,x3,y3, width=0.05, color='blue')
plt.arrow(x0,y0,x4,y4, width=0.05, color='blue')

plt.plot(x,y, 'or')
plt.text(x + 0.2, y, 'A')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid()
print(x,y)

print(get_regular_coord(2*np.sqrt(3), -np.sqrt(3), np.pi/6))
```
