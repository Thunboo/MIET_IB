```python
# ВАРИАНТ №1
# 2)
# x^2 - 2xy + y^2 - 10x - 6y + 25 = 0
# Вычисляя своим пешком приходим к вывод, что:
# ALPHA = PI / 4 

# x = x' cos(phi) - y' sin(phi) = x'/sqrt(2) - y'/sqrt(2)
# y = x' sin(phi) + y' cos(phi) = x'/sqrt(2) + y'/sqrt(2)

# Каноническое уравнение параболы: (y - sqrt(2)/2)^2 = 4*sqrt(2)*(x - 3*sqrt(2)/2)

import numpy as np
import matplotlib.pyplot as plt

angle = np.pi/4

y = np.linspace(-10, 10, 100)
x = (( y - np.sqrt(2)/2 )**2 + 12) / 4*np.sqrt(2)

plt.plot(x, y)
plt.arrow(-10,0,18,0,width=0.15)
plt.arrow(0,-10,0,18,width=0.15)
plt.arrow(-5,-5,10,10,width=0.15, color='green')
plt.arrow(-5, 5, 10, -10,width=0.15, color='green')

plt.plot( x*np.cos(angle) - y*np.sin(angle), x*np.sin(angle) + y*np.cos(angle))

plt.grid()
plt.axis('equal')

'''
Угол поворота: 45 градусов
Базис старой кривой: 
    i{1/sqrt(2), -1/sqrt(2)}
    j{1/sqrt(2),  1/sqrt(2)}
'''
```
