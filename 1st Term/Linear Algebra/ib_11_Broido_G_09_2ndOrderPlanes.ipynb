```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sympy import *
```

```python
# x^2 + 2y^2 = z
x = np.linspace(-1,1,200)
y = np.linspace(-1,1,200)

[x,y] = np.meshgrid(x,y) # СОЗДАЁМ СЕТКУ
z = lambda x,y: x**2 + 2*y**2
Z = z(x,y)

fig = plt.figure( figsize=(10,10) )
ax = plt.axes(projection='3d')

my_cmap = plt.get_cmap('magma')

surface = ax.plot_surface(x,y,Z, rstride=18, cstride=8, alpha=0.8, cmap=my_cmap) 
# rstride и cstride - размер разбиения на мелкие квадраты
# alpha - opacity
cset = ax.contourf(x, y, Z, zdir='z', offset = np.min(Z), cmap = my_cmap)
cset = ax.contourf(x, y, Z, zdir='x', offset = -2, cmap = my_cmap)
cset = ax.contourf(x, y, Z, zdir='y', offset = 2, cmap = my_cmap)
# ПРОЕКЦИИ
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

ax.view_init(30,30) # POV
```

```python
# ПОСТРОЕНИЕ ЭЛЛИПСОИДА
# x^2 /9  +  y^2 /1  +  z^2 /1 = 1

fig = plt.figure( figsize=(10,10))
ax = plt.axes(projection='3d')

coefs = (9,1,1)
rx,ry,rz = np.sqrt(coefs)

tetha = np.linspace( -np.pi/2, np.pi/2, 200)
phi = np.linspace( 0, 2*np.pi, 200)

x = rx * np.outer( np.cos(phi), np.cos(tetha)) # Все возможные произведения cos на cos
y = ry * np.outer( np.sin(phi), np.cos(tetha))
z = rz * np.outer( np.ones_like(phi), np.sin(tetha)) # искусственно заполненый столбец для соразмерности в матрицу при помощи np.outer 

my_cmap = plt.get_cmap('rainbow')
surface = ax.plot_surface(x,y,z, rstride=10, cstride=10, alpha=0.6, cmap = my_cmap)
```

```python
# x^2 + y^2 = 1 
# ЦИЛИНДР

fig = plt.figure( figsize=(10,10))
ax = plt.axes(projection='3d')

x = np.linspace(-1,1,30)
z = np.linspace(0,3,30)
x,z = np.meshgrid(x,z)

y = np.sqrt(1 - x**2)

ax.plot_wireframe(x, y,z, color='b')
ax.plot_wireframe(x,-y,z, color='b')

ax.view_init(60,45) # POV
```

```python
# № 2
# Тор
# ПОСТРОЕНИЕ ЭЛЛИПСОИДА
# x^2 /9  +  y^2 /1  +  z^2 /1 = 1

fig = plt.figure( figsize=(10,10))
ax = plt.axes(projection='3d')

coefs = (1,1,1)

rx,ry,rz = np.sqrt(coefs)

tetha = np.linspace(0, 2*np.pi, 200)
phi   = np.linspace(0, 2*np.pi, 200)

x = np.outer( 2*rx + ry * np.cos(phi), np.cos(tetha)) # Все возможные произведения cos на cos
y = np.outer( 2*rx + ry * np.cos(phi), np.sin(tetha))
z =  ry * np.outer(np.sin(phi), np.ones_like(tetha)) # искусственно заполненый столбец для соразмерности в матрицу при помощи np.outer 

my_cmap = plt.get_cmap('rainbow')
surface = ax.plot_surface(x,y,z, rstride=10, cstride=10, alpha=0.6, cmap = my_cmap)

plt.axis('equal')
ax.view_init(40,80)
```
