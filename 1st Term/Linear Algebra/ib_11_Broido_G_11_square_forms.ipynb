```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

```python
# Переход от одного базиса к другому
# vector * matrix => new_vector in new basis
# Ax = x' 
# ( 1 2 )
# ( 3 4 )
 
# ( 1 - l 2    )
# ( 3     4 - l) 

# | 1 - l 2    |
# | 3     4 - l| 
# => Quadratic equation
# => l1, l2 = ... - Собственные числа матрицы

'''
( l1  0 )
(  0 l2 ) 
# Диагональная форма
B новом базисе
'''
```

```python
# (1,3; 4,2)
A = np.array([[1,3], [4,2]])
L, T = np.linalg.eig(A)
print(L) # СОБСТВЕННЫЕ ЧИСЛА
print(T) # СОБСТВЕННЫЕ ВЕКТОРА - BASIS !!!
# МАТРИЦА ПЕРЕХОДА (как часть - матрица поворота)
# D = T^-1 * A * T
D = np.linalg.inv(T).dot(A).dot(T)
print(D) # Главная диагональ - СОБСТВЕННЫЕ ЧИСЛА
AA = Matrix([[1,3],[4,2]])
T,D = AA.diagonalize()
print()
print(T,'\n',D)
```

```python
# 3x^2 + 10xy + y^2
A = Matrix([[3,10//2], [10//2,3]])
T, D = A.diagonalize()
x, y = symbols('x y')
f = D[0,0] * x**2 + D[1,1] * y**2
print(f) # Квадратичная форма в базисе из собственных векторов
```

```python
# 6x^2 + 5y^2 + 7z^2 - 4xy + 4xz
A = Matrix([[ 6,-2, 2], # x
            [-2, 5, 0], # y
            [ 2, 0, 7]])# z
#             x  y  z
T, D = A.diagonalize()
print(T)
print(D)
```

```python
# 3x^2 + 10xy + 3y^2 - 32
x1, y1 = symbols('x1 y1')
X1 = Matrix([[x1, y1]])
AA = Matrix([[3,10//2],
             [10//2, 3]])
T, D = AA.diagonalize()
print(T)
print(D)
X = T/sqrt(2) * X1.T # Чтоб такой же длины был базис => НОРМИРОВАННАЯ МАТРИЦА ПЕРЕХОДА
                     # Без транспанированной не работает, это же строка)
#print(X)
print(X[0,0])
print('x =', simplify(X[0,0]))
print('y =', simplify(X[1,0]))
print()

F = 3*X[0]**2 + 10*X[0]*X[1] + 3*X[1]**2 - 32
print(simplify(F)/32)

var('x y')
plot_implicit(Eq(3*x**2 + 10*x*y + 3*y**2, 32))
plot_implicit(Eq(-x**2/16     +     y**2/4, 1))
```

```python
# вариант 1
# 2)
# x^2 - 2xy + y^2 - 10x - 6y + 25 = 0
# (x - 5)^2 - 25 + (y - 3)^2 - 9 - 2xy + 25 = 0
# (x - 5)^2 + (y - 3)^2 - 9 - 2xy = 0
# x1^2 + y1^2 - 9 - 2(x1 + 5)(y1 + 3)


# 2(x - 5)(y - 3) = 2(xy - 5y - 3x + 15) = 2xy - 10y - 6x + 30
# 2xy = 2(x-5)(y-3) + 10y + 6x - 30

###############################################
# 3x^2 + 10xy + 3y^2 - 32

# x^2 - 2xy + y^2 - 10x - 6y + 25 = 0
x1, y1 = symbols('x1 y1')
X1 = Matrix([[x1, y1]])
AA = Matrix([[ 1,-1],
             [-1, 1]])
T, D = AA.diagonalize()
print(T)
print(D)

X = T/sqrt(2) * X1.T # Чтоб такой же длины был базис => НОРМИРОВАННАЯ МАТРИЦА ПЕРЕХОДА
                     # Без транспанированной не работает, это же строка)

print(X)

print(X[0,0])
print('x =', simplify(X[0,0]))
print('y =', simplify(X[1,0]))
print()

F = 1*X[0]**2 - 2*X[0]*X[1] + 1*X[1]**2 - 10*X[0] - 6*X[1] + 25
print(simplify(F)/25)

var('x y')
plot_implicit(Eq(1*x**2 - 2*x*y + 1*y**2 - 10*x - 6*y, 25))
plot_implicit(Eq(-8*np.sqrt(2)*x/25 + 2*y**2/25 + 2*np.sqrt(2)*y/25, 1))

################################################
```
