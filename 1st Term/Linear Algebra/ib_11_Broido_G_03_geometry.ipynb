```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

L = Line( (1,4), (5,3) )
print(L.equation()) # Общее уравнение прямой
print(L.arbitrary_point()) # x = ... ; y = ... 
# ^^^^^^^ Параметрическое задание прямой
print('\n')

# (-2, 3) k = 2
A = Point(-2, 3)
k = 2
L = Line(A, slope = k)
print(L.equation())
```

```python
# 5x - 4y - 8 = 0
# [-5; 5]
# (0; -2) нарисовать вектор нормали
# орт вектора нормали <=> 

x = np.linspace(-5,5, 100)
n = [5, -4]
ortN = n / np.linalg.norm(n)

plt.subplot()
plt.plot(x, (5*x - 8) / 4)
plt.arrow(0, -2, n[0], n[1], width = 0.05, color = 'Red')
plt.arrow(0, 0, ortN[0], ortN[1], width = 0.05, color = 'Green')
plt.axis('equal')
plt.title('МОЙ ШИКАРНЫЙ РИСУНОК')
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
```

```python
# НАЙТИ РАССТОЯНИЕ ОТ ТОЧКИ ДО ПРЯМОЙ
# (-1, 1)
# (1, 4), (5, 3)
L = Line( (1,4), (5,3) )
P = Point(-1, 1)
print(L.distance(P))
```

```python
# НАЙТИ УГОЛ МЕЖДУ ПРЯМЫМИ (НОРМАЛЯМИ) (через скалярн. пр-ие)
s = Line( (0,0) , (1,0) )
w = Line( (1,1) , (0,0) )
s1 = w.angle_between(s)
s2 = w.smallest_angle_between(s)
print(s1)
print(s2)
```

```python
# Дана точка и прямая. Через точку построить параллельную прямую
# 144 из учебника
# -2x + y - 1 = 0
# (-1, 2)
s = Line( (0,1), (-0.5, 0) )
p = Point(-1, 2)
w = s.parallel_line( p )
print(w.equation() * 2)
```

```python
# Построить прямую, перпеникулярную этой через точку
w = s.perpendicular_line( p )
print(w.equation() * 2)
```

```python
# Найти проекцию точки напрямую
w = s.projection( p )
print(w)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

A = Point(2, -2)
B = Point(3,  5)
C = Point(6,  1)

AC = Line(A, C)
BH = AC.perpendicular_line( B )
print(BH.equation())
H = BH.points[1]
d = B.distance( H )
print("Length of height =", d)

M = [ (B[0] + C[0])/2 , (B[1] + C[1])/2 ]
AM = Line(A, M)
print(AM.equation() * 2)
print("Length of AM =", A.distance( M ))

AB = Line(A, B)
CK = AB.perpendicular_line( C )
print(CK.equation())
K = CK.points[1]
d = C.distance( K )
print("Length of CK =", d)

plt.plot(A[0], A[1], 'or', B[0], B[1], 'og', C[0], C[1], 'ob')
plt.arrow(int(A[0]), int(A[1]), int(C[0] - A[0]), int(C[1] - A[1]),
           width = 0.05, length_includes_head = True, color='k')
plt.arrow(int(A[0]), int(A[1]), int(B[0] - A[0]), int(B[1] - A[1]),
           width = 0.05, length_includes_head = True, color='k')
plt.arrow(int(B[0]), int(B[1]), int(C[0] - B[0]), int(C[1] - B[1]),
           width = 0.05, length_includes_head = True, color='k')
plt.arrow(int(A[0]), int(A[1]), float(M[0] - A[0]), float(M[1] - A[1]),
           width = 0.05, length_includes_head = True, color='g')
plt.arrow(int(C[0]), int(C[1]), float(K[0] - C[0])/2, float(K[1] - C[1])/2,
           width = 0.05, length_includes_head = True, color='c')
plt.plot(M[0], M[1], 'oy')

plt.text(A[0] + 0.2, A[1] - 0.2, '$A$')
plt.text(B[0] + 0.2, B[1] - 0.2, '$B$')
plt.text(C[0] + 0.2, C[1] - 0.2, '$C$')
plt.text(C[0] - 0.6, C[1] - 0.1, '$H$')
plt.text(M[0] + 0.2, M[1] - 0.2, '$M$')
plt.text((K[0] + C[0])/2 - 0.4, (K[1]+ C[1])/2 - 0.2, '$K$')
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('equal')
plt.show()

print("Заметим, что треугольник в нашей задаче является прямоугольным-равнобедренным!",
      "\nBH - высота, совпадает с BC, тк угол C - прямой угол",
      "\nAM - медиана",
      "\nCK - биссектриса")
```
